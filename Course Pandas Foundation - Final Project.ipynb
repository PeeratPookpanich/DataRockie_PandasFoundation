```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
```

```python
# filter 2017 data
df_2017 = df[df['Order Date'].dt.year == 2017]
df_2017
```

```python
# total sales
total_sales = df_2017['Sales'].sum()
print(f'total sales = {total_sales}')
```

```python
# average sales
avg_sales = df_2017['Sales'].mean()
print(f'average sales = {avg_sales}')
```

```python
# Standard deviation of sales
sd_sales = df_2017['Sales'].std()
print(f'standard deviation of sales = {sd_sales}')
```

```python
# TODO 06 - which Segment has the highest profit in 2018
```

```python
df[df['Order Date'].dt.strftime('%Y') == '2018'].groupby('Segment')['Profit'].sum()
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
```

```python
df[(df['Order Date'] >= '2019-04-15') & (df['Order Date'] <= '2019-12-31')].groupby('State')['Sales'].sum().sort_values(ascending=False).tail(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
```

```python
# filter West + Central Region
west_central2019 = df[df['Order Date'].dt.year == 2019]

# the proportion of total sales
proportion_sales = west_central2019.query('Region == "West" | Region == "Central"')['Sales'].sum()*100/west_central2019['Sales'].sum()
proportion_sales
print(f'the proportion of total sales (%) in West + Central in 2019 is {proportion_sales:.2f} %')
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
```

```python
df2020= df[(df['Order Date'].dt.year == 2019)|(df['Order Date'].dt.year == 2020)]

top10_product = df2020.groupby('Product Name')[['Quantity','Sales']].sum().reset_index().sort_values('Quantity', ascending = False).head(10)
top10_product
```
